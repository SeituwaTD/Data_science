```python
# Importation of libraries
import pandas as pd
import numpy as np
```

```python
#loading of files and merging the both of them
df1 = pd.read_csv("QVI_purchase_behaviour.csv")
df2 = pd.read_excel("QVI_transaction_data.xlsx")
df1.head(5)
df2.head(5)
```

```python
df1.head(5)
```

```python
# I will merge on lylty_card_nbr since that is the column the two datasets have in common
df = pd.merge(df1,df2,on="LYLTY_CARD_NBR")
df.head(5)
```

```python
# Exploratory function
def explore_data(df):
    print("Shape of dataset:",df.shape)
    print("\nColumns: \n", df.columns)
    print("\nData types:\n", df.dtypes)
    print("\nMissing values:\n", df.isnull().sum())
    print("\nSummary statistics:\n", df.describe())

```

```python
explore_data(df)
```

```python
# let us clean our data futher
def clean_data(df):
    # Drop duplicates
    df= df.drop_duplicates()
    # Handle missing values if any although our data has none
    df= df.fillna(method='ffill')
    return df
```

```python
clean_data(df)
```

```python
df.groupby(['LYLTY_CARD_NBR','PREMIUM_CUSTOMER','LIFESTAGE','PROD_NAME', 'PROD_QTY'])['TOT_SALES'].sum().sort_values(ascending=False)
```
