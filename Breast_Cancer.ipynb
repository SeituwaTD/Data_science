# Dataset is complete and has no missing values

```python
# Map diagnosis column with 1 representing malignant and 0 representing benign
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
df.head()
```

```python
df = df.drop(['id'], axis=1)
df.head()
```

```python
df['diagnosis'].value_counts()
```

```python
# correlation analysis
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.show()
```

```python
diagnosis_corr = df.corr()['diagnosis'].sort_values(ascending=True)
print(diagnosis_corr)
```

```python
# spliting data into training and testing datasets
x = df.drop('diagnosis', axis=1)
y = df['diagnosis']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)
```

```python
# Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Comments
- From the evaluation metrics, the model can be said to be somewhat above average.
- The models gives 70 true negative, 1 false positive, 3 false negative and 1 true positive.
- In medicine false negative is seen as the most dangerous error.
- This model gives out false negative results in approximately 3% of the time which should be improved on.
